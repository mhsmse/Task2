```python
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

df = pd.read_csv("No-FMCG.csv")


importance_cols = [f'F-{i}' for i in range(1, 16)]


df_clean = df.dropna(subset=importance_cols)

# Calculate the average importance score for each feature
feature_avg_importance = df_clean[importance_cols].mean().sort_values(ascending=False)

# Visualize the importance of each feature
plt.figure(figsize=(10, 6))
sns.barplot(x=feature_avg_importance.index, y=feature_avg_importance.values, palette="viridis")
plt.title("Average Feature Importance")
plt.xlabel("Features")
plt.ylabel("Average Importance Score")
plt.xticks(rotation=45)
plt.show()
# Calculate the average importance score for each feature
feature_avg_importance = df_clean[importance_cols].mean()

# Sort the features by average importance score in descending order
feature_avg_importance = feature_avg_importance.sort_values(ascending=False)

# Report the average importance score for each feature
print("Average Importance Score for Each Feature:")
print(feature_avg_importance)
```

```python
# Calculate the average importance score for each feature
feature_avg_importance = df_clean[importance_cols].mean().sort_values(ascending=False)

# Perform ANOVA to check for significant differences between features
f_stat, p_value = stats.f_oneway(*[df_clean[col] for col in importance_cols])

# Print the results
print("\nANOVA Test Results:")
print(f"F-statistic: {f_stat:.3f}, p-value: {p_value:.5f}")
if p_value < 0.05:
    print("Significant differences exist between feature importance scores.")
else:
    print("No significant differences detected between feature importance scores.")
```
